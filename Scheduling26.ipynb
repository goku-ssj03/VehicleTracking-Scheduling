```python
def no_of_buses(n):
  bus_depart_time=[]
  if (n<=12): n=n
  else: n=12
  for j in range (0,n):
    m=n/12
    a=int(60/n)
    b=int(j*a/5)
    c=5*(b)
    bus_depart_time.append(c)
  return bus_depart_time
```

```python
df1=pd.DataFrame()
info_of_routes=[]
for j in range(0,n):
  for k in range(0,n):
    if(j!=k):

      freq=np.random.randint(10,size=19)
      hour1=np.random.randint(2)
      min1=np.random.randint(60)
      r_time=dt.time(hour1,min1)

      hour2=0
      min2=np.random.randint(30)
      b_time=dt.time(hour2,min2)

      r_id=100*(ord(points[j]))+10*(ord(points[k]))+ord((points[j]))-ord(points[k])

      l1=[]
      for i in range (0,19):
        a=f'{dt.time(i+4)}-{dt.time(i+5)}'
        l1.append(a)
      df001=pd.DataFrame([freq],index=range(1),columns=range(19))
      df001.columns=l1


      dict_of_route={
      "route_id":[r_id],
      "running_time":[r_time],
      "break_time":[b_time],
      "source":[points[j]],
      "destination":[points[k]],
      }

      info_of_routes.append(dict_of_route)
      df11=pd.DataFrame(dict_of_route)
      df11=pd.concat([df11,df001],axis=1)
      df1=pd.concat([df1,df11])
df1=df1.reset_index(drop=True)
df1
```

```python
def sour(point):
  f1_out=df1[(df1['source']==point)]
  f1_out=f1_out.reset_index(drop=True)
  return(f1_out)
```

```python
def min(a):
    return ((int(a[0:2])*60)+(int(a[3:5])))
```

```python
def no_of_buses(n):
  bus_depart_time=[]
  if (n<=12): n=n
  else: n=12
  for j in range (0,n):
    m=n/12
    a=int(60/n)
    b=int(j*a/5)
    c=5*(b)
    bus_depart_time.append(c)
  return bus_depart_time
```

```python
df201=pd.DataFrame()
total_n_initial=[100]
main_trip_arr=[0]
n_initial_pointwise=[]

running_bus=[]

for minute in range(240,1380,5):

  index=(int(minute/60)+1)

  for j in range (0,n):
    n_initial_at_point=[]
    df152=sour(points[j])

    for k in range (0 ,n):
      if j!=k:
        if df152['destination'].isin([points[k]]).any():

          start_place=points[j]
          end_place=points[k]

          row_index=df152.loc[df152['destination']==end_place].index[0]

          m=df152.iat[row_index,index]
          m=int(m)
          bus_depart_time=no_of_buses(m)

          if (minute - 60*(int (minute/60))) in bus_depart_time:

            for i in range (0,len(n_initial_pointwise)):
              if 'place' in n_initial_pointwise[i]:
                if (n_initial_pointwise[i]['place'][0]==points[j] and n_initial_pointwise[i]['ready_for_next'][0]<=minute):
                  n_initial_at_point.append(n_initial_pointwise[i]['id'][0])
                  n_initial_pointwise[i].clear()
                  break

            if (len(n_initial_at_point)==0):
              n_initial_at_point.append(total_n_initial[-1]+1)
              main_trip_arr.append(0)
              total_n_initial.append(n_initial_at_point[-1])

            curr_bus=n_initial_at_point[0]
            n_initial_at_point.pop(0)
            main_trip_arr[curr_bus-100]+=1

            r_t_dt=df152.iat[row_index,1]
            r_t=(str(r_t_dt))
            b_t_dt=df152.iat[row_index,2]
            b_t=str(b_t_dt)
            r_id=df152.iat[row_index,0]
            red_t=min(r_t)+min(b_t)+minute
            if (red_t>=1440):
              red_t=red_t-1440
            dict1={
            "id":[curr_bus],
            "trip_no":[main_trip_arr[(curr_bus-100)]],
            "route_id":[r_id],
            "source":[points[j]],
            "time_of_depart":[dt.time(int(minute/60),minute-(60*int(minute/60)))],
            "run_time":[r_t],
            "break":[b_t],
            "ready_for_next":[dt.time(int(red_t/60),red_t-(60*int(red_t/60)))],
            "destination":[points[k]]
            }

            running_bus.append(dict1)
            df202=pd.DataFrame(dict1)
            df201=pd.concat([df201,df202])

            awesome = [ minute - 4, minute -3, minute -2, minute -1, minute]

            for i in range (0,len(running_bus)):
              if 'destination' in running_bus[i]:
                if (running_bus[i]['destination'][0]==points[j]):
                  if (min(str(running_bus[i]['ready_for_next'][0]))) in awesome:
                    stored_id = running_bus[i]['id'][0]
                    curr_place= running_bus[i]['destination'][0]
                    reaching_time=running_bus[i]['ready_for_next'][0]

                    running_bus[i].clear()

                    dict2={
                    "id": [stored_id],
                    "place":[curr_place],
                    "ready_for_next":[min(str(reaching_time))]
                    }
                    n_initial_pointwise.append(dict2)


df201["index"]=list(np.arange(1,len(df201)+1))
df201


```
